```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

folds_path = '/root/cv_project/data/train_folds.csv'
folds = pd.read_csv(folds_path)
folds.columns = folds.columns.str.strip()

# fold 컬럼 자동 탐색
fold_col = [c for c in folds.columns if 'fold' in c.lower()]
fold_col = fold_col[0] if fold_col else None

# 라벨 컬럼 자동 탐색
label_col = [c for c in folds.columns if c.lower() in ['label', 'target', 'class']]
label_col = label_col[0] if label_col else None

if fold_col is None:
    raise ValueError(f"⚠️ fold 관련 컬럼이 없습니다. 현재 컬럼: {list(folds.columns)}")

if label_col is None:
    raise ValueError(f"⚠️ label/target 컬럼을 찾을 수 없습니다. 현재 컬럼: {list(folds.columns)}")

print(f"✅ fold 컬럼: {fold_col}")
print(f"✅ label 컬럼: {label_col}")

# fold별 라벨 분포 시각화
plt.figure(figsize=(8,4))
sns.countplot(x=fold_col, hue='target', data=folds)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Label Distribution per Fold')
plt.tight_layout()
plt.show()
```
